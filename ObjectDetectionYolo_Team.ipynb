## Tahap 5: Evaluation

Setelah proses pelatihan model selesai, langkah selanjutnya adalah menilai kinerja model untuk memastikan bahwa hasil yang diperoleh memenuhi kebutuhan analisis.

```python
paths = [
    ('/content/object-detection/train2/results.csv', 'YOLOv8n'),
    ('/content/object-detection-v9/train/results.csv', 'YOLOv9t'),
    ('/content/object-detection/train3/results.csv', 'YOLOv10n'),
    ('/content/object-detection/train4/results.csv', 'YOLOv11n'),
]

for path, model_name in paths:
    df_results = pd.read_csv(path)
    plt.figure(figsize=(10, 6))
    plt.plot(df_results.index, df_results['train/box_loss'], label='Box Loss')
    plt.plot(df_results.index, df_results['train/cls_loss'], label='Class Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'Training Loss per Epoch - {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

Membaca data hasil pelatihan dari tiga file CSV yang masing-masing merepresentasikan model YOLOv8, YOLOv9, YOLOv10, dan YOLOv11. Untuk setiap model, data loss (kerugian) dikaji dan divisualisasikan dalam bentuk grafik. Grafik menampilkan nilai box loss dan class loss terhadap jumlah epoch (iterasi pelatihan), dengan tujuan untuk memantau performa pelatihan setiap model.

```python
metrics1 = model1.val(split='test')
metrics2 = model2.val(split='test')
metrics3 = model3.val(split='test')
metrics4 = model4.val(split='test')
```

Mengevaluasi kinerja masing-masing model (model1, model2, model3, dan model4) menggunakan data uji (test split), dan hasil evaluasi disimpan dalam variabel metrics1, metrics2, metrics3, dan metrics4.

```python
def metricsF1(metrics):
    precision = metrics.results_dict.get('metrics/precision(B)', 0)
    recall = metrics.results_dict.get('metrics/recall(B)', 0)
    if precision + recall != 0:
        f1_score = 2 * (precision * recall) / (precision + recall)
    else:
        f1_score = 0

    return f1_score
```

Fungsi metricsF1 menghitung skor F1 berdasarkan nilai precision dan recall yang diperoleh dari hasil evaluasi model. Jika precision dan recall tidak nol, skor F1 dihitung menggunakan rumus harmonik tapi jika salah satunya nol, skor F1 diatur menjadi nol.

```python
df = pd.DataFrame({
    'Model': ['YOLOv8n', 'YOLOv9t', 'YOLOv10n', 'YOLOv11n'],
    'Precision': [
        round(metrics1.results_dict["metrics/precision(B)"], 4),
        round(metrics2.results_dict["metrics/precision(B)"], 4),
        round(metrics3.results_dict["metrics/precision(B)"], 4),
        round(metrics4.results_dict["metrics/precision(B)"], 4)
    ],
    'Recall': [
        round(metrics1.results_dict["metrics/recall(B)"], 4),
        round(metrics2.results_dict["metrics/recall(B)"], 4),
        round(metrics3.results_dict["metrics/recall(B)"], 4),
        round(metrics4.results_dict["metrics/recall(B)"], 4)
    ],
    'F1 Score': [
        metricsF1(metrics1),
        metricsF1(metrics2),
        metricsF1(metrics3),
        metricsF1(metrics4)
    ],
    'mAP50': [
        round(metrics1.results_dict["metrics/mAP50(B)"], 4),
        round(metrics2.results_dict["metrics/mAP50(B)"], 4),
        round(metrics3.results_dict["metrics/mAP50(B)"], 4),
        round(metrics4.results_dict["metrics/mAP50(B)"], 4)
    ]
})

df = df.round(4)
df
```

Menyusun hasil evaluasi dari tiga model deteksi objek (YOLOv8n, YOLOv9t, YOLOv10n, dan YOLOv11n) ke dalam sebuah dataframe dengan menggunakan pustaka pandas. Metrik yang dimasukkan meliputi Precision, Recall, mAP50, dan mAP50-95. Nilai-nilai metrik diambil dari atribut results_dict hasil evaluasi masing-masing model. Selanjutnya, nilai pada dataframe dibulatkan hingga tiga angka di belakang koma untuk mempermudah pembacaan hasil evaluasi.

```python
plotMatrics = [
    ("mAP50", "Perbandingan mAP@0.5"),
    ("Precision", "Perbandingan Precision"),
    ("Recall", "Perbandingan Recall"),
    ("F1 Score", "Perbandingan F1 Score")
]

for metric, title in plotMatrics:
    plt.figure(figsize=(10, 6))
    plt.bar(df['Model'], df[metric], color = ['skyblue', 'lightgreen', 'salmon', 'plum'])
    plt.title(title)
    plt.ylabel(metric)
    plt.ylim(0, 1)
    plt.grid(True)
    plt.show()
```

Membuat visualisasi perbandingan empat metrik evaluasi model deteksi objek, yaitu mAP@0.5, mAP@0.5:0.95, Precision, dan Recall. Untuk setiap metrik, dibuat grafik batang (bar chart) yang menampilkan nilai metrik dari masing-masing model (YOLOv8n, YOLOv9t, YOLOv10n, dan YOLOv11n). Grafik diberi judul sesuai dengan jenis metrik yang dibandingkan, serta disesuaikan agar skala sumbu y berada dalam rentang 0 hingga 1. Visualisasi ini bertujuan untuk mempermudah analisis performa antar model.

```python
def predict_BoundingBox(img_path, results):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = box.conf[0].item()
        cls = int(box.cls[0].item())

        label = f"{model3.names[cls]}: {conf:.2f}"
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    return img
```

Menampilkan hasil prediksi deteksi objek pada sebuah gambar. Fungsi ini menerima path gambar (img_path) dan hasil deteksi (results). Gambar dibaca dan dikonversi ke format RGB, kemudian untuk setiap bounding box hasil deteksi, digambar kotak hijau pada area objek terdeteksi, lengkap dengan label yang berisi nama kelas dan nilai confidence. Label diambil berdasarkan daftar kelas pada model3. Fungsi mengembalikan gambar yang telah diberi anotasi deteksi.

```python
prediksi = model4.predict(source='/content/dataset/test/images', save=False)
```

Menjalankan proses prediksi menggunakan model model3 terhadap seluruh gambar yang terdapat pada direktori /content/dataset/test/images. Parameter save=False menunjukkan bahwa hasil prediksi tidak disimpan ke dalam file, melainkan hanya disimpan sementara di memori. Hasil prediksi kemudian disimpan dalam variabel prediksi, yang dapat digunakan untuk analisis atau visualisasi lebih lanjut.

```python
for i, result in enumerate(prediksi[:2]):
    img_path = result.path
    img_with_boxes = predict_BoundingBox(img_path, result)

    plt.figure(figsize=(8, 6))
    plt.imshow(img_with_boxes)
    plt.axis("off")
    plt.title(f"Gambar {i+1}")
    plt.show()
```

Menampilkan dua hasil prediksi pertama dari model deteksi objek model3. Untuk setiap hasil prediksi, diambil path gambar asli, kemudian fungsi predict_BoundingBox digunakan untuk menambahkan anotasi bounding box dan label pada gambar. Gambar yang telah dianotasi ditampilkan menggunakan matplotlib tanpa sumbu koordinat, serta diberi judul sesuai nomor urutan (Gambar 1 dan Gambar 2). Visualisasi ini bertujuan untuk melihat hasil deteksi objek secara langsung pada gambar uji.

```python
#up gambar
uploaded = files.upload()
img_name = list(uploaded.keys())[0]

# Buka gambar dengan PIL
image = Image.open(io.BytesIO(uploaded[img_name]))
image = np.array(image)

# Resize gambar ke 640x640
image_resized = cv2.resize(image, (640, 640))
image_resized = cv2.cvtColor(image_resized, cv2.COLOR_RGB2BGR)

# Simpan gambar
temp_img_path = "temp_input.jpg"
cv2.imwrite(temp_img_path, image_resized)

# Prediksi
user_result = model2(temp_img_path)

image_with_boxes = predict_BoundingBox(temp_img_path, user_result[0])
image_with_boxes = cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8, 6))
plt.imshow(image_with_boxes)
plt.axis("off")
plt.title("Hasil Prediksi Gambar Input User")
plt.show()
```

Melakukan prediksi deteksi objek pada gambar yang diunggah oleh pengguna. Gambar diunggah melalui antarmuka Colab dan dibaca menggunakan PIL, lalu diubah ke format array dan di-resize ke ukuran 640x640 piksel sesuai dengan kebutuhan model YOLO. Setelah dikonversi ke format warna BGR, gambar disimpan sementara sebagai temp_input.jpg. Model deteksi objek kemudian digunakan untuk memprediksi objek dalam gambar tersebut.
