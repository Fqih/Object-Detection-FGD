## Tahap 5: Evaluation

Setelah model dilatih, tahap selanjutnya adalah evaluasi untuk menilai kinerja model terhadap data uji. Pada tahap ini kami membandingkan beberapa metrik performa seperti class_loss, box_loss, precision, recall, F1 score, dan mAP50 antara model-model yang telah dilatih.

```python
paths = [
    ('/content/runs/detect/train/results.csv', 'YOLOv8n'),
    ('/content/runs/detect/train2/results.csv', 'YOLOv9t'),
    ('/content/runs/detect/train3/results.csv', 'YOLOv10n'),
    ('/content/runs/detect/train4/results.csv', 'YOLOv11n'),
]

class_losses = []
box_losses = []
model_names = []

for path, model_name in paths:
    df_results = pd.read_csv(path)
    class_losses.append(df_results['train/cls_loss'])
    box_losses.append(df_results['train/box_loss'])
    model_names.append(model_name)
```

Penjelasan: Mendefinisikan path ke file CSV yang berisi hasil pelatihan untuk masing-masing model YOLO yang telah dilatih. Hasil pelatihan ini akan digunakan untuk mengukur kinerja masing-masing model. Kemudian membaca file CSV hasil pelatihan dan mengumpulkan data class_loss dan box_loss untuk masing-masing model. Data ini digunakan untuk menganalisis kinerja model dalam hal pengenalan objek dan pengaturan bounding box.

```python
plt.figure(figsize=(10, 6))
for i, model_name in enumerate(model_names):
    plt.plot(df_results.index, class_losses[i], label=f'{model_name} Class Loss')
plt.xlabel('Epoch')
plt.ylabel('Class Loss')
plt.title('perbandingan class loss dari masing masing yolo')
plt.legend()
plt.grid(True)
plt.show()
```

Penjelasan: Membuat grafik perbandingan class_loss dari masing-masing model selama proses pelatihan untuk melihat seberapa cepat masing-masing model dapat mengurangi kesalahan klasifikasi.

```python
plt.figure(figsize=(10, 6))
for i, model_name in enumerate(model_names):
    plt.plot(df_results.index, box_losses[i], label=f'{model_name} Box Loss')
plt.xlabel('Epoch')
plt.ylabel('Box Loss')
plt.title('Box Loss per Epoch for Each YOLO Model')
plt.legend()
plt.grid(True)
plt.show()
```

Penjelasan: Membuat grafik untuk perbandingan box_loss antara model-model YOLO, yang menunjukkan bagaimana kesalahan dalam mendeteksi posisi bounding box berubah selama pelatihan.

```python
metrics1 = model1.val(split='test')
metrics2 = model2.val(split='test')
metrics3 = model3.val(split='test')
metrics4 = model4.val(split='test')
```

Penjelasan: Mengevaluasi setiap model pada data uji dan menyimpan hasil evaluasi dalam variabel masing-masing, yang mencakup metrik seperti precision, recall, dan mAP50.

```python
df = pd.DataFrame({
    'Model': ['YOLOv8n', 'YOLOv9t', 'YOLOv10n', 'YOLOv11n'],
    'Precision': [
        round(metrics1.results_dict["metrics/precision(B)"], 4),
        round(metrics2.results_dict["metrics/precision(B)"], 4),
        round(metrics3.results_dict["metrics/precision(B)"], 4),
        round(metrics4.results_dict["metrics/precision(B)"], 4)
    ],
    'Recall': [
        round(metrics1.results_dict["metrics/recall(B)"], 4),
        round(metrics2.results_dict["metrics/recall(B)"], 4),
        round(metrics3.results_dict["metrics/recall(B)"], 4),
        round(metrics4.results_dict["metrics/recall(B)"], 4)
    ],
    'mAP': [
        round(metrics1.results_dict["metrics/mAP50(B)"], 4),
        round(metrics2.results_dict["metrics/mAP50(B)"], 4),
        round(metrics3.results_dict["metrics/mAP50(B)"], 4),
        round(metrics4.results_dict["metrics/mAP50(B)"], 4)
    ]
})

df = df.round(4)
df
```

Penjelasan: Menyusun hasil evaluasi untuk setiap model dalam bentuk tabel, yang mencakup nilai precision, recall, F1 score, dan mAP50, sehingga kita dapat membandingkan kinerja setiap model.

```python
plotMatrics = [
    ("mAP", "Perbandingan mAP@0.5"),
    ("Precision", "Perbandingan Precision"),
    ("Recall", "Perbandingan Recall"),
]

for metric, title in plotMatrics:
    plt.figure(figsize=(10, 6))
    plt.bar(df['Model'], df[metric], color = ['skyblue', 'lightgreen', 'salmon', 'plum'])
    plt.title(title)
    plt.ylabel(metric)
    plt.ylim(0, 1)
    plt.grid(True)
    plt.show()
```

Penjelasan: Membuat grafik batang untuk membandingkan metrik evaluasi (mAP50, Precision, Recall) antar model. Setiap grafik memberikan perbandingan visual untuk memudahkan analisis.

```python
def predict_BoundingBox(img_path, results):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = box.conf[0].item()
        cls = int(box.cls[0].item())

        label = f"{model3.names[cls]}: {conf:.2f}"
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    return img
```

Penjelasan: Fungsi ini digunakan untuk membaca gambar, mengonversinya ke RGB, dan menggambar bounding box serta label pada gambar berdasarkan hasil deteksi. Gambar yang sudah diproses kemudian dikembalikan.

```python
prediksi = model4.predict(source='/content/dataset/test/images', save=False)
```

Penjelasan: Melakukan prediksi pada gambar yang ada di folder /content/dataset/test/images menggunakan model model4 (YOLOv11n). Hasil prediksi tidak disimpan karena argumen save=False.

```python
for i, result in enumerate(prediksi[10:12]):
    img_path = result.path
    img_with_boxes = predict_BoundingBox(img_path, result)

    plt.figure(figsize=(8, 6))
    plt.imshow(img_with_boxes)
    plt.axis("off")
    plt.title(f"Gambar {i+1}")
    plt.show()
```

Penjelasan: Menampilkan hasil prediksi untuk gambar ke-10 dan ke-11 dari hasil prediksi. Setiap gambar diproses untuk menggambar bounding box, kemudian ditampilkan dengan judul yang sesuai menggunakan matplotlib.

```python
# Upload gambar
uploaded = files.upload()
img_name = list(uploaded.keys())[0]

# Buka dan konversi gambar
image = Image.open(io.BytesIO(uploaded[img_name]))
image = np.array(image)
image_resized = cv2.resize(image, (640, 640))
image_resized = cv2.cvtColor(image_resized, cv2.COLOR_RGB2BGR)

# Simpan gambar sementara
temp_img_path = "temp_input.jpg"
cv2.imwrite(temp_img_path, image_resized)

# Fungsi prediksi + waktu inferensi
def predict_with_timing(model, img_path):
    start = time.time()
    results = model(img_path)
    end = time.time()
    duration = round(end - start, 4)
    return results, duration

# Asumsi model1 s.d. model4 sudah didefinisikan (YOLOv8 sampai YOLOv11)
models = [model1, model2, model3, model4]
model_names = ["YOLOv8", "YOLOv9", "YOLOv10", "YOLOv11"]

# Plot hasil prediksi semua model
plt.figure(figsize=(20, 10))
for i, (mdl, name) in enumerate(zip(models, model_names), 1):
    res, infer_time = predict_with_timing(mdl, temp_img_path)
    image_with_boxes = predict_BoundingBox(temp_img_path, res[0])
    image_with_boxes = cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 2, i)
    plt.imshow(image_with_boxes)
    plt.axis("off")
    plt.title(f"{name} â€” {infer_time:.4f} detik")

plt.suptitle("Perbandingan Hasil Prediksi dan Waktu Inferensi", fontsize=16)
plt.tight_layout()
plt.show()
```

Penjelasan: Mengunggah gambar, mengubah ukuran dan konversi warna gambar, lalu menyimpannya sementara. Selanjutnya, fungsi predict_with_timing digunakan untuk menjalankan prediksi dengan model YOLO (dari YOLOv8 hingga YOLOv11) dan mengukur waktu inferensi. Hasil prediksi dari setiap model ditampilkan dalam plot bersama dengan waktu inferensi masing-masing.
